# Label Encoding
##### Also called an OrdinalEncoder
##### In this encoding, each category is assigned a value from 1 through N (here N is the number of categories for the feature. One major issue with this approach is there is no relation or order between these classes, but the algorithm might consider them as some order, or there is some relationship.

```python
from sklearn.preprocessing import LabelEncoder
df['Ctg_label']=LabelEncoder().fit_transform(df.Category)
df
```

```python
#Pandas factorize also perform the same function
df.loc[:,'Ctg_factorized_encode']=pd.factorize(df['Category'])[0].reshape(-1,1)
df
```

# Ordinal Encoding
##### This encoding looks almost similar to Label Encoding but slightly different as Label coding would not consider whether variable is ordinal or not and it will assign sequence of integers
##### -as per the order of data (Pandas assigned Ex: Hot (0), Cold (1), “Very Hot” (2) and Warm (3)) or
##### -as per alphabetical sorted order (scikit-learn assigned Ex: Cold(0), Hot(1), “Very Hot” (2) and Warm (3)).

```python
import warnings
warnings.filterwarnings("ignore")
C_dict={ 'ART_AND_DESIGN':1,
          'AUTO_AND_VEHICLES':2,
           'BEAUTY':3
}
df['Ctg_Ordinal']=df.Category.map(C_dict)
df.head(100)
```

# Helmert Encoding
##### In this encoding, the mean of the dependent variable for a level is compared to the mean of the dependent variable over all previous levels.

# Binary Encoding

```python
#If there are n unique categories,then binary encoding results in the only log(base 2)ⁿ features
#for 100 categories One Hot Encoding will have 100 features while for Binary encoding, we will need just seven features
import category_encoders as ce #We will use the category_encoders package
encoder=ce.BinaryEncoder(cols=['Category']) #and the function name is BinaryEncoder
dfbin=encoder.fit_transform(df['Category'])
df=pd.concat([df,dfbin],axis=1)
df
```

# Frequency Encoding

##### It is a way to utilize the frequency of the categories as labels. In the cases where the frequency is related somewhat with the target variable, it helps the model to understand and assign the weight in direct and inverse proportion, depending on the nature of the data. Three-step for this :
##### -Select a categorical variable you would like to transform
##### -Group by the categorical variable and obtain counts of each category
##### -Join it back with the training dataset

```python
len(df)
```

```python
df.groupby('Category').size()
```

```python
fe=df.groupby('Category').size()/len(df)
df.loc[:,'Cat_freq_encode']=df.Category.map(fe)
df
```

```python
data={'Temperature':['Hot',"Cold","Very Hot","Warm","Hot","Warm","Warm","Hot","Hot","Cold"],
      'Color':["Red","Yellow","Blue","Blue",'Red',"Yellow","Red","Yellow","Yellow","Yellow"],
       'Target':[1,1,1,0,1,0,1,0,1,1]
    }
df1=pd.DataFrame(data,columns=['Temperature','Color','Target'])
df1
```

# Mean Encoding or Target Encoding

```python
#The advantages of the mean target encoding are that it does not affect the volume of the data and helps in faster learning
mean_encode=df1.groupby('Temperature')['Target'].mean()
print(mean_encode)
df1.loc[:,'Temp_mean_enc']=df1['Temperature'].map(mean_encode)
df1
```

```python
mean_encode=df.groupby('Category')['Rating'].mean()
print(mean_encode)
df.loc[:,'Cat_mean_enc']=df['Category'].map(mean_encode)
df
```

```python
#There are many variations of this target encoding in practice, like smoothing:
#Compute the golbal mean
mean=df1['Target'].mean()
#Compute the number of values and the meanof each group
agg=df1.groupby('Temperature')['Target'].agg(['count','mean'])
counts=agg['count']
means=agg['mean']
weight=100
#Compute the smoothed means
smooth=(counts*means+weight*mean)/(counts+weight)
#Replace each value by the according smoothed mean
print(smooth)
df1.loc[:,'Temp_smean_enc']=df1['Temperature'].map(smooth)
df1
```

```python
mean=df['Rating'].mean()
agg=df.groupby('Category')['Rating'].agg(['count','mean'])
counts=agg['count']
means=agg['mean']
weight=100
smooth=(counts*means+weight*mean)/(counts+weight)
print(smooth)
df.loc[:,'Cat_smean_enc']=df['Category'].map(smooth)
df
```

# Weight of Evidence Encoding 

####  Weight of evidence (WOE) is a measure of how much the evidence supports or undermines a hypothesis

![](img.jpg) WoE will be 0 if the P(Goods) / P(Bads) = 1. That is if the outcome is random for that group. If P(Bads) > P(Goods) the odds ratio will be < 1 and the WoE will be < 0; if, on the other hand, P(Goods) > P(Bads) in a group, then WoE > 0.

```python
#We calculate probability of target=1 i.e. Good=1 for each category 
woe_df=df1.groupby('Temperature')['Target'].mean()
woe_df=pd.DataFrame(woe_df)
#Rename the column name to 'Good' to keep it consistent with formula for easy unerstanding
woe_df=woe_df.rename(columns={'Target':'Good'})
#Calculate Bad probability which is 1- Good probability
woe_df['Bad']=1-woe_df.Good
#We need to add a small value to avoid divide by zero in denominatory
woe_df['Bad']=np.where(woe_df['Bad']==0,0.000001,woe_df['Bad'])
#compute the Woe
woe_df['WoE']=np.log(woe_df.Good/woe_df.Bad)
woe_df
```

```python
#Map the WOE values backto each row of DataFrame
df1['WoE_Encode']=df1['Temperature'].map(woe_df['WoE'])
df1
```

```python
woe_df=df.groupby('Category')['Rating'].mean()
woe_df=pd.DataFrame(woe_df)

woe_df=woe_df.rename(columns={'Rating':'Good'})

woe_df['Bad']=woe_df.Good.max()-woe_df.Good

woe_df['Bad']=np.where(woe_df['Bad']==0,0.000001,woe_df['Bad'])

woe_df['WOE']=np.log(woe_df.Good/woe_df.Bad)
woe_df
df['WOE_Encode']=df['Category'].map(woe_df['WOE'])
df
```

# Probability Ratio Encoding
##### For each label, we calculate the mean of target=1, that is the probability of being 1 ( P(1) ), and also the probability of the target=0 ( P(0) ). And then, we calculate the ratio P(1)/P(0) and replace the labels by that ratio.

```python
#We calculate probability of target=1 i.e. Good=1 for each category 
pr_df=df1.groupby('Temperature')['Target'].mean()
pr_df=pd.DataFrame(pr_df)
#Rename the column name to 'Good' to keep it consistent with formula for easy unerstanding
pr_df=pr_df.rename(columns={'Target':'Good'})
#Calculate Bad probability which is 1- Good probability
pr_df['Bad']=1-pr_df.Good
#We need to add a small value to avoid divide by zero in denominatory
pr_df['Bad']=np.where(pr_df['Bad']==0,0.000001,pr_df['Bad'])
#compute the Probability Ratio
pr_df['PR']=pr_df.Good/pr_df.Bad
pr_df
```

```python
#Map the Probability Ratio values back to each row of DataFrame
df1['PR_Encode']=df1['Temperature'].map(pr_df['PR'])
df1
```

```python
pr_df=df.groupby('Category')['Rating'].mean()

pr_df=pd.DataFrame(pr_df)

pr_df=pr_df.rename(columns={'Rating':'Good'})

pr_df['Bad']=pr_df.Good.max()-pr_df.Good

pr_df['Bad']=np.where(pr_df['Bad']==0,0.000001,pr_df['Bad'])

pr_df['PR']=pr_df.Good/pr_df.Bad

df['PR_Encode']=df['Category'].map(pr_df['PR'])

df
```
